```python
# Importing Important Libraries 
import pandas as pd
import numpy as nu 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# Reading the dataset
dataset=pd.read_csv("E:\ML\AM_dataset\AM\cleaned_data_lpbf.csv")
```

```python
# Finding any duplicates
duplicate_rows = dataset[dataset.duplicated()]
duplicate_rows.count()
```

```python
# Display duplicate rows
print("Duplicate Rows except first occurrence:")
print(duplicate_rows)
```

```python
# Remove duplicates and update the DataFrame
dataset_no_duplicates = dataset.drop_duplicates()

# dataset without duplicates
print("DataFrame without duplicates:")
print(dataset_no_duplicates)
```

```python
dataset.shape
dataset1=dataset.copy()
```

```python
dataset_no_duplicates.head()
dataset2=dataset_no_duplicates
```

```python
dataset2.head()
```

```python
dataset2.shape
```

```python
# Finding correlations between features
plt.figure(figsize=(10,6))
sns.heatmap(dataset2.corr(),annot=True,cmap="magma",fmt=".2f",linewidth=0.5)
plt.title("Features Correlations Heatmap")
plt.show()
```

```python
dataset2.describe()
```

```python
# Pairplot
sns.pairplot(dataset2)
```

```python

# Selecting input features
input_features = dataset2.iloc[:, :-3]

# Scatter plots for each input feature
num_features = input_features.shape[1]

# Set up subplots
fig, axes = plt.subplots(nrows=num_features, ncols=1, figsize=(7, 5 * num_features))

# Iterate through each input feature
for i, column in enumerate(input_features.columns):
    # Create a legend based on unique material values
    for material_value in dataset2["Material"].unique():
        subset_data = dataset2[dataset2["Material"] == material_value]
        sc = axes[i].scatter(x=subset_data["Yield strength (MPa)"], y=subset_data[column], cmap='viridis', alpha=0.7, label=f"{material_value}")

    axes[i].set_xlabel("Yield Strength (MPa)")
    axes[i].set_ylabel(f"{column}")
    axes[i].set_title(f"{column} vs Yield Strength")
    axes[i].grid(True)
    axes[i].legend()

plt.tight_layout()
plt.show()
```
